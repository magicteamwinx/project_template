6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Cylinder(Model3D):
   def __init__(self, radius, height):
        self.radius = radius
        self.height = height
   def get_volume(self):
        # π * r² * h
        return math.pi * (self.radius ** 2) * self.height

class Cube(Model3D):
   def __init__(self, side):
        self.side = side
   def get_volume(self):
        return self.side ** 3

my_cube = Cube(3)
my_cylinder = Cylinder(2, 5)

total = my_cube + my_cylinder

print(f"Cube volume: {my_cube.get_volume()}")
print(f"Cylinder volume: {my_cylinder.get_volume():.2f}")
print(f"Total volume: {total:.2f}")
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, wallet_id, initial_balance=0.0):
        # Приватні атрибути з подвійним підкресленням
        self.__id = wallet_id
        self.__balance = float(initial_balance)

    # Геттер для ID (тільки для читання, без сеттера)
    @property
    def wallet_id(self):
        return self.__id

    # Геттер для балансу
    @property
    def balance(self):
        return self.__balance

    # Сеттер для балансу (контролює логіку зміни)
    @balance.setter
    def balance(self, value):
        if value < 0:
            print("Помилка: Баланс не може бути від'ємним!")
        else:
            self.__balance = value

    # Метод для поповнення
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount  # Використовує сеттер
            print(f"Поповнено на {amount}. Поточний баланс: {self.balance}")
        else:
            print("Сума поповнення має бути додатною!")

    # Метод для зняття
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount  # Використовує сеттер
            print(f"Знято {amount}. Поточний баланс: {self.balance}")
        elif amount > self.balance:
            print("Недостатньо коштів на рахунку!")
        else:
            print("Сума зняття має бути додатною!")

# --- Демонстрація роботи ---
my_wallet = EWallet("UA-777", 100)

print(f"Гаманець: {my_wallet.wallet_id}") # Звертаємось як до атрибута
my_wallet.deposit(50)                    # Поповнюємо
my_wallet.withdraw(30)                   # Знімаємо
my_wallet.balance = -500                 # Спроба прямого запису через сеттер (спрацює перевірка)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name}: {self.price} грн"


class ShoppingCart:
    def __init__(self):
        # Список для зберігання об'єктів типу Item
        self.items = []

    def add_item(self, item):
        """Додає об'єкт Item до кошика"""
        if isinstance(item, Item):
            self.items.append(item)
            print(f"Додано: {item.name}")
        else:
            print("Помилка: Можна додавати лише об'єкти класу Item")

    def remove_item(self, item_name):
        """Видаляє айтем з кошика за його назвою"""
        for item in self.items:
            if item.name == item_name:
                self.items.remove(item)
                print(f"Видалено: {item_name}")
                return
        print(f"Товар '{item_name}' не знайдено в кошику")

    def show_cart(self):
        """Відображає всі товари в кошику та загальну вартість"""
        if not self.items:
            print("Кошик порожній.")
            return

        print("\n--- Ваш кошик ---")
        total_price = 0
        for item in self.items:
            print(f"- {item}")
            total_price += item.price
        print(f"Загальна сума: {total_price} грн\n")


# --- Демонстрація роботи ---

# Створюємо товари
apple = Item("Яблуко", 15)
milk = Item("Молоко", 40)
bread = Item("Хліб", 25)

# Створюємо кошик
my_cart = ShoppingCart()

# Додаємо товари
my_cart.add_item(apple)
my_cart.add_item(milk)
my_cart.add_item(bread)

# Показуємо вміст
my_cart.show_cart()

# Видаляємо один товар
my_cart.remove_item("Молоко")

# Фінальний список
my_cart.show_cart()
```

10. (13б). Створіть клас Vehicle з атрибутами brand (manufacturer) та production_age. Створіть ще один клас Engine з такими атрибутами, як type (ex. petrol) та size (volume, litres). Створіть клас Car, який успадковує обидва класи Vehicle та Engine. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Car та отримавши mro для цього класу.

```python
class Vehicle:
    def __init__(self, brand, production_age):
        self.brand = brand
        self.production_age = production_age

    def display_vehicle_info(self):
        print(f"Brand: {self.brand}, Age: {self.production_age} years")

class Engine:
    def __init__(self, engine_type, size):
        self.engine_type = engine_type
        self.size = size

    def display_engine_info(self):
        print(f"Engine: {self.engine_type}, Volume: {self.size}L")

# Клас Car успадковує обидва класи одночасно
class Car(Vehicle, Engine):
    def __init__(self, brand, production_age, engine_type, size):
        # Ініціалізуємо обидва батьківські класи
        Vehicle.__init__(self, brand, production_age)
        Engine.__init__(self, engine_type, size)

    def drive(self):
        print(f"The {self.brand} is driving with its {self.size}L engine!")

# --- Демонстрація ---

# Створюємо об'єкт класу Car
my_car = Car("BMW", 3, "petrol", 2.0)

# Використовуємо методи обох батьківських класів
my_car.display_vehicle_info()
my_car.display_engine_info()
my_car.drive()

# Отримуємо MRO (Method Resolution Order)
print("\nMethod Resolution Order (MRO) для класу Car:")
for cls in Car.mro():
    print(cls)
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekIterator:
    def __init__(self):
        self.days = [
            "Понеділок", "Вівторок", "Середа",
            "Четвер", "Пʼятниця", "Субота", "Неділя"
        ]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration


week = WeekIterator()

for day in week:
    print(day)

another_week = WeekIterator()
print(f"\nПерший день: {next(another_week)}")
```
